```python
# Import
import ansys.fluent.core as pyfluent
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

print("All libraries imported successfully")
print(f"PyFluent version: {pyfluent.__version__}")
```

```python
# Launch Fluent
solver = pyfluent.launch_fluent(
    product_version="261",
    mode="solver",
    show_gui=False,
    dimension=2,        # 2D simulation
    precision="double"  # double precision
)

print("Fluent launched successfully")
```

```python
# Define design points
design_points = [
    {"dp": "dp0", "angle": 0,  "cas": r"dp0\FFF\Fluent\FFF.1-4.cas.h5",  "dat": r"dp0\FFF\Fluent\FFF.1-4-00106.dat.h5"},
    {"dp": "dp1", "angle": 4,  "cas": r"dp1\FFF\Fluent\FFF-5.cas.h5",    "dat": r"dp1\FFF\Fluent\FFF-5-00042.dat.h5"},
    {"dp": "dp2", "angle": 8,  "cas": r"dp2\FFF\Fluent\FFF-5.cas.h5",    "dat": r"dp2\FFF\Fluent\FFF-5-00043.dat.h5"},
    {"dp": "dp3", "angle": 12, "cas": r"dp3\FFF\Fluent\FFF-5.cas.h5",    "dat": r"dp3\FFF\Fluent\FFF-5-00054.dat.h5"},
    {"dp": "dp4", "angle": 16, "cas": r"dp4\FFF\Fluent\FFF-5.cas.h5",    "dat": r"dp4\FFF\Fluent\FFF-5-00500.dat.h5"},
]
```

```python
import os

# Check file sizes to see if they are identical
base_path = r"C:\Users\enriq\OneDrive - UCB-O365\ANSYS\ANSYS Projects\Airfoil\Airfoil Simulation_files"

files_to_check = [
    r"dp0\FFF\Fluent\FFF.1-4-00106.dat.h5",
    r"dp1\FFF\Fluent\FFF.1-3-00107.dat.h5",
    r"dp2\FFF\Fluent\FFF.1-3-00107.dat.h5",
    r"dp3\FFF\Fluent\FFF.1-3-00107.dat.h5",
    r"dp4\FFF\Fluent\FFF.1-3-00107.dat.h5",
]

for f in files_to_check:
    full_path = f"{base_path}\\{f}"
    size = os.path.getsize(full_path)
    print(f"{f.split(chr(92))[0]} — {size:,} bytes")
```

```python
results = []

for dp in design_points:
    try:
        print(f"\n{'='*55}")
        print(f"  Processing {dp['dp']} — Angle of Attack: {dp['angle']}°")
        print(f"{'='*55}")

        # Load case
        print("  Loading case file...")
        solver.settings.file.read(
            file_type="case",
            file_name=f"{base_path}\\{dp['cas']}"
        )

        # Load data
        print("  Loading data file...")
        solver.settings.file.read(
            file_type="data",
            file_name=f"{base_path}\\{dp['dat']}"
        )

        # Delete existing report definitions cleanly
        try:
            solver.settings.solution.report_definitions.lift["lift-force"].delete()
        except:
            pass
        try:
            solver.settings.solution.report_definitions.drag["drag-force"].delete()
        except:
            pass

        # Recreate lift report definition
        solver.settings.solution.report_definitions.lift["lift-force"] = {}
        solver.settings.solution.report_definitions.lift["lift-force"].set_state({
            "zones": ["aerofoil"],
            "force_vector": [0, 1, 0]
        })

        # Recreate drag report definition
        solver.settings.solution.report_definitions.drag["drag-force"] = {}
        solver.settings.solution.report_definitions.drag["drag-force"].set_state({
            "zones": ["aerofoil"],
            "force_vector": [1, 0, 0]
        })

        # Compute forces
        print("  Extracting lift and drag via PyFluent...")
        lift_result = solver.settings.solution.report_definitions.compute(
            report_defs=["lift-force"]
        )
        drag_result = solver.settings.solution.report_definitions.compute(
            report_defs=["drag-force"]
        )

        lift = lift_result[0]["lift-force"][0]
        drag = drag_result[0]["drag-force"][0]
        ld   = lift / drag if drag != 0 else 0

        results.append({
            "angle_deg":    dp["angle"],
            "lift_N":       lift,
            "drag_N":       drag,
            "lift_to_drag": ld
        })

        print(f"  ✓ Lift: {lift:.6f} N | Drag: {drag:.6f} N | L/D: {ld:.4f}")

    except Exception as e:
        print(f"  ✗ Skipping {dp['dp']} — {e}")
```

```python
# Build dataframe and save CSV
df = pd.DataFrame(results)
print("\n" + "="*55)
print("  RESULTS SUMMARY")
print("="*55)
print(df.to_string(index=False))

df.to_csv("data/airfoil_results.csv", index=False)
print("\n✓ Results saved to data/airfoil_results.csv")
```

```python
# ── Generate plots ────────────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle("NACA 0012 Airfoil — CFD Parametric Analysis\nExtracted via PyAnsys (PyFluent) | ANSYS Fluent 2026 R1", 
             fontsize=13, fontweight="bold")

# Plot 1 — Lift curve
axes[0].plot(df["angle_deg"], df["lift_N"], "b-o", linewidth=2, markersize=8)
axes[0].set_xlabel("Angle of Attack (°)", fontsize=11)
axes[0].set_ylabel("Lift Force (N)", fontsize=11)
axes[0].set_title("Lift Curve", fontsize=12)
axes[0].grid(True, alpha=0.3)
axes[0].fill_between(df["angle_deg"], df["lift_N"], alpha=0.1, color="blue")

# Plot 2 — Drag curve
axes[1].plot(df["angle_deg"], df["drag_N"], "r-o", linewidth=2, markersize=8)
axes[1].set_xlabel("Angle of Attack (°)", fontsize=11)
axes[1].set_ylabel("Drag Force (N)", fontsize=11)
axes[1].set_title("Drag Curve", fontsize=12)
axes[1].grid(True, alpha=0.3)
axes[1].fill_between(df["angle_deg"], df["drag_N"], alpha=0.1, color="red")

# Plot 3 — Lift to drag ratio
axes[2].plot(df["angle_deg"], df["lift_to_drag"], "g-o", linewidth=2, markersize=8)
axes[2].set_xlabel("Angle of Attack (°)", fontsize=11)
axes[2].set_ylabel("L/D Ratio", fontsize=11)
axes[2].set_title("Lift-to-Drag Ratio", fontsize=12)
axes[2].grid(True, alpha=0.3)
axes[2].fill_between(df["angle_deg"], df["lift_to_drag"], alpha=0.1, color="green")

# Annotate peak L/D
max_ld_idx = df["lift_to_drag"].idxmax()
axes[2].annotate(
    f"Peak L/D: {df['lift_to_drag'][max_ld_idx]:.2f}\nat {df['angle_deg'][max_ld_idx]}°",
    xy=(df["angle_deg"][max_ld_idx], df["lift_to_drag"][max_ld_idx]),
    xytext=(df["angle_deg"][max_ld_idx] + 1, df["lift_to_drag"][max_ld_idx] - 1.5),
    fontsize=9,
    arrowprops=dict(arrowstyle="->", color="black")
)

plt.tight_layout()
plt.savefig("outputs/airfoil_analysis.png", dpi=150, bbox_inches="tight")
plt.show()
print("✓ Plot saved to outputs/airfoil_analysis.png")
```
